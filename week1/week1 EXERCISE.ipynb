```python
# imports
import os
from openai import OpenAI
from scraper import fetch_website_contents
from dotenv import load_dotenv
```

```python
API_KEY = os.getenv('OPENAI_API_KEY')
```

```python
# constants

MODEL_GPT = 'gpt-5-mini'
MODEL_LLAMA = 'llama3.2'
openai = OpenAI()
```

```python
def Book_Summary_GPT(url):
    summary = fetch_website_contents(url)

    user_prompt = "this is the website to be scraped: " + summary

    system_prompt = "You are a helpful assistant that will receive the scraped content of a book. Using the content, please provide a short summary of the contents of the book"
    
    response = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {'role':'system', 'content': system_prompt},
            {'role':'user', 'content':user_prompt}
        ]
    )
    print(response.choices[0].message.content)
    
```

```python
# Get gpt-4o-mini to answer, with streaming
Book_Summary_GPT('https://sobrief.com/books/the-subtle-art-of-not-giving-a-fuck')
```

```python
def Book_Summary_Ollama(url):
    summary = fetch_website_contents(url)

    user_prompt = "this is the website to be scraped: " + summary

    system_prompt = "You are a helpful assistant that will receive the scraped content of a book. Using the content, please provide a short summary of the contents of the book"


    ollama = OpenAI(base_url="http://localhost:11434/v1", api_key='ollama')

    response = ollama.chat.completions.create(model='phi3', messages=[
        {'role': 'system', 'content':system_prompt},
        {'role':'user', 'content':user_prompt}
    ])

    print(response.choices[0].message.content)

```

```python
Book_Summary_GPT(url='https://sobrief.com/books/1984')

```
